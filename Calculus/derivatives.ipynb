# Derivatives

## Definition of the Derivative

The limit definition of the derivative of $f(x)$ is 

$$ f'(x) = \frac{d(f(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h} $$

For the derivative of $f(x)$ at a specific point—in this example, $x=a$—the derivative is defined as

$$ f'(a) = \frac{d(f(a))}{dx} = \lim_{x \rightarrow a} \frac{f(x) - f(a)}{x-a} $$

## Addition rule

$$ \frac{d(f(x) + g(x))}{dx} = f'(x) + g'(x) $$

### Proof

Start with the first limit definition of the derivative:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) + g(x+h) - (f(x) + g(x))}{h} $$

Distribute the negative sign:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) + g(x+h) - f(x) - g(x)}{h} $$

Rearrange the terms:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) - f(x) + g(x+h) - g(x)}{h} $$

Split the fraction into two terms:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \biggl( \frac{f(x+h) - f(x)}{h} + \frac{g(x+h) - g(x)}{h} \biggr) $$

The limit of a sum is the sum of the limits:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h} + \lim_{h \rightarrow 0} \frac{g(x+h) - g(x)}{h} $$

The left-hand term is $f'(x)$ and the right-hand term is $g'(x)$:

$$ \frac{d(f(x) + g(x))}{dx} = f'(x) + g'(x) $$

## Product rule

$$ \frac{d(f(x) \cdot g(x))}{dx} = f'(x) g(x) + f(x) g'(x) $$

### Proof

Start with the first limit definition of the derivative:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) g(x+h) - f(x) g(x)}{h} $$

Add zero by subtracting and then adding $f(x+h)g(x)$ in the numerator:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) g(x+h) - f(x+h)g(x) + f(x+h)g(x) - f(x) g(x)}{h} $$

Split the fraction into two terms:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \biggl( \frac{f(x+h) g(x+h) - f(x+h)g(x)}{h} + \frac{f(x+h)g(x) - f(x) g(x)}{h} \biggr) $$

The limit of a sum is the sum of the limits:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} \frac{f(x+h) g(x+h) - f(x+h)g(x)}{h} + \lim_{h \rightarrow 0} \frac{f(x+h)g(x) - f(x) g(x)}{h} $$

In the left-hand term, pull out a factor of $f(x+h)$; in the right-hand term, pull out a factor of $g(x)$:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} f(x+h) \biggl( \frac{g(x+h) - g(x)}{h} \biggr) + \lim_{h \rightarrow 0} g(x) \biggl( \frac{f(x+h) - f(x)}{h} \biggr) $$

In the right-hand term, since $g(x)$ is not dependent on $h$, we can pull it out of the limit:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} f(x+h) \biggl( \frac{g(x+h) - g(x)}{h} \biggr) + g(x) \lim_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h} $$

The right-hand limit is $f'(x)$:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} f(x+h) \biggl( \frac{g(x+h) - g(x)}{h} \biggr) + g(x) f'(x) $$

For the left-hand term, the limit of a product is equal to the product of the limits of each factor:

$$ \frac{d(f(x) + g(x))}{dx} = \lim_{h \rightarrow 0} f(x+h) \cdot \lim_{h \rightarrow 0} \frac{g(x+h) - g(x)}{h} + g(x) f'(x) $$

The limit of the first factor is $f(x)$, and the limit of the second factor is $g'(x)$:

$$ \frac{d(f(x) + g(x))}{dx} = f(x) g'(x) + g(x) f'(x) $$

Rearrange the second term to get the initial statement:

$$ \frac{d(f(x) \cdot g(x))}{dx} = f'(x) g(x) + f(x) g'(x) $$

## Chain rule

Given two functions, $f(x)$ and $g(x)$, where $g(x)$ is the input of $f(x)$ (i.e., $f(g(x)$) and $f'(x) = \frac{d(f(x))}{dx}$ and $g'(x) = \frac{d(g(x))}{dx}$, then

$$ \frac{d(f(g(x))}{dx} = f'(g(x)) \cdot g'(x) $$

### Proof

Start with the limit definition for the derivative of $f(g(x))$:

$$ \frac{d}{dx}(f(g(x)) = \lim_{h \rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{h} $$

Multiply the numerator and the denominator by $g(x+h) - g(x)$:

$$ \frac{d}{dx}(f(g(x)) = \lim_{h \rightarrow 0} \frac{(f(g(x+h)) - f(g(x))) \cdot (g(x+h) - g(x))}{h \cdot (g(x+h) - g(x))} $$

Rearrange the denominator:

$$ \frac{d}{dx}(f(g(x)) = \lim_{h \rightarrow 0} \frac{(f(g(x+h)) - f(g(x))) \cdot (g(x+h) - g(x))}{(g(x+h) - g(x)) \cdot h} $$

Split the fraction into the product of two fractions:

$$ \frac{d}{dx}(f(g(x)) = \lim_{h \rightarrow 0} \biggl( \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)} \cdot \frac{g(x+h) - g(x)}{h} \biggr) $$

The limit of a product is equal to the product of the limits:

$$ \frac{d}{dx}(f(g(x)) = \lim_{h \rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)} \cdot \lim_{h \rightarrow 0} \frac{g(x+h) - g(x)}{h} $$

The limit of the right-hand term is the derivative of $g(x)$:

$$ \frac{d}{dx}(f(g(x)) = \lim_{h \rightarrow 0} \frac{f(g(x+h)) - f(g(x))}{g(x+h) - g(x)} \cdot g'(x) $$

Let's make a substitution.  Let $x_1 = g(x+h)$ and let $a = g(x)$. Note that as $h \rightarrow 0$, $x_1 \rightarrow g(x)$, which equals $a$:

$$ \frac{d}{dx}(f(g(x)) = \lim_{x_1 \rightarrow a} \frac{f(x_1) - f(a)}{x_1 - a} \cdot g'(x) $$

The left-hand term now has the same form as the derivative definition at a specific point, meaning the left-hand term is the derivative of $f(x)$ evaluated at $a$, and $a = g(x)$:

$$ \frac{d}{dx}(f(g(x)) = f'(g(x)) \cdot g'(x) $$

The chain rule is also used to take derivatives in terms of other variables, where it is written as

$$ \frac{dy}{dx} = \frac{dy}{du} \frac{du}{dx} $$

## Quotient rule

$$ \frac{d}{dx} \biggl( \frac{f(x)}{g(x)} \biggr) = \frac{g(x)f'(x) - f(x)g'(x)}{(g(x))^2} $$

### Proof

This requires using the product rule, and you might notice that the chain rule makes an appearance. First, change the fraction into a product:

$$ \frac{d}{dx} \biggl( \frac{f(x)}{g(x)} \biggr) = \frac{d(f(x) \cdot g(x)^{-1})}{dx} $$

Make a u-substitution.  Let $u = g(x), \frac{du}{dx} = g'(x)$:

$$ \frac{d}{dx} \biggl( \frac{f(x)}{g(x)} \biggr) = \frac{d(f(x) \cdot u^{-1})}{dx} $$

Using the product rule, we get

$$ \frac{d(f(x) \cdot u^{-1})}{dx}  = f'(x) \cdot u^{-1} + f(x) \cdot (-u)^{-2} \cdot \frac{du}{dx} $$

Rewrite this as fractions:

$$ \frac{d(f(x) \cdot u^{-1})}{dx}  = \frac{f'(x)}{u} - \frac{f(x)}{u^2} \cdot \frac{du}{dx} $$

Add the fractions together with a common denominator:

$$ \frac{d(f(x) \cdot u^{-1})}{dx}  = \frac{u f'(x) - f(x) \frac{du}{dx}}{u^2} $$

Replace $u$ with $g(x)$ and $\frac{du}{dx}$ with $g'(x)$:

$$ \frac{d}{dx} \biggl( \frac{f(x)}{g(x)} \biggr) = \frac{g(x)f'(x) - f(x)g'(x)}{(g(x))^2} $$

The rhyme that is often used to remember the order of the terms is 
"Low dee high less high dee low, draw the line and square below",
meaning the bottom function ("low") times the derivative ("dee") of the top function ("high"), minus ("less") 
the top function times the derivative of the bottom function, all divided by ("draw the line") the bottom function squared ("square below").
